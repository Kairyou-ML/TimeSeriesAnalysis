## ARIMA

Estimating ARIMA model 1


```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = pd.read_csv("TSA_DATA\Data01_02_FoodstuffCPI.csv", parse_dates=True)
CPI_rate = data['CPI_rate']
```

```python
from statsmodels.tsa.arima.model import ARIMA
X = CPI_rate.values
X = X.astype('float32')
```

```python
# Fit model
model = ARIMA(X, order=(1, 1, 1))
model_fit = model.fit()

import matplotlib.pyplot as pyplot
 
# summary of fit model
print(model_fit.summary())
# line plot of residuals
residuals = pd.Series(model_fit.resid)
residuals.plot()
pyplot.show()
# density plot of residuals
residuals.plot(kind='kde')
pyplot.show()
# summary stats of residuals
print(residuals.describe())

```

Forecasting ARIMA model


Mean values Forecastin

```python
STEPS = 20
forecasts_df = model_fit.get_forecast (steps=STEPS).summary_frame()

df1 = CPI_rate
ax = df1.plot(figsize=(12, 6))
plt.ylabel('CPI_rate')
forecasts_df['mean'].plot(style='k--')
ax.fill_between(
forecasts_df.index,
forecasts_df['mean_ci_lower'],
forecasts_df['mean_ci_upper'],
color='k',
alpha=0.1
)
```

Split data series to train part and test part


```python
train = CPI_rate[CPI_rate.index <= 80]
test = CPI_rate[CPI_rate.index > 80]

plt.plot(train, color = "black")
plt.plot(test, color = "red")
plt.ylabel('CPI_rate')
plt.xlabel('Time')
plt.xticks(rotation=45)
plt.title("Train/Test split for CPI rate Data")
plt.show()
```

```python
ARIMAmodel = ARIMA(CPI_rate, order = (2, 0, 1))
ARIMAmodel = ARIMAmodel.fit()

y_pred = ARIMAmodel.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 
y_pred_df["Predictions"] = ARIMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_out = y_pred_df["Predictions"] 
plt.plot(y_pred_out, color='red', label = 'ARIMA Predictions')
plt.plot(test, color='Yellow', label = 'Testing')
plt.plot(train, color='black', label = 'Train')
plt.legend()
```

Evaluate the accuracy of forecasting


```python
import numpy as np
from sklearn.metrics import mean_squared_error

arma_rmse = np.sqrt(mean_squared_error(test.values, y_pred_df["Predictions"]))
print("RMSE: ",arma_rmse)
```

## Seasonal ARIMA (SARIMA) Model


ThegeneralnotationforaSARIMAmodelisexpressedas:
SARIMA(p,d,q)(P,D,Q, m). Where:
- p:Orderofthenon-seasonalautoregressivepart.
- d:Degreeofdifferencing(non-seasonal).
- q:Orderofthenon-seasonalmovingaveragepart.
- P:Orderoftheseasonalautoregressivepart.
- D:Degreeofseasonaldifferencing.
- Q:Orderoftheseasonalmovingaveragepart.
- m:Numberofperiodsineachseason
      (e.g.,12formonthlydatawithyearlyseasonality).


```python
data = pd.read_csv("TSA_DATA\Data01_02_FoodstuffCPI.csv", parse_dates=True)
CPI_rate = data['CPI_rate']
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

SARIMAXmodel = SARIMAX(CPI_rate, order = (2, 0, 1), seasonal_order=(2,0,1,12))
SARIMAXmodel = SARIMAXmodel.fit()

y_pred = SARIMAXmodel.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 
y_pred_df["Predictions"] = SARIMAXmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_out = y_pred_df["Predictions"] 
plt.plot(test, color='Yellow', label = 'Testing')
plt.plot(train, color='black', label = 'Train')
plt.plot(y_pred_out, color='Blue', label = 'SARIMA Predictions')
plt.legend()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

arma_rmse = np.sqrt(mean_squared_error(test.values, y_pred_df["Predictions"]))
print("RMSE: ",arma_rmse)
```

```python
import pandas as pd

df = pd.read_csv("TSA_DATA\Data01_02_FoodstuffCPI.csv")

df['Date'] = pd.to_datetime(dict(year=df['Year'], month=df['Month'], day=1))
df = df.set_index('Date')

# Series 1: Foodstuff CPI, Series 3: CPI_rate
ts1 = df['Foodstuff_toDecCPI']
ts3 = df['CPI_rate']

import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.plot(ts1, label='Foodstuff CPI')
plt.plot(ts3, label='CPI Rate')
plt.legend(); plt.title("Foodstuff CPI & CPI Rate (Monthly)"); plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller, kpss

def stationarity_tests(series, name="Series"):
    print(f"\n=== Stationarity tests for {name} ===")
    adf_res = adfuller(series.dropna(), autolag='AIC')
    print(f"ADF Statistic: {adf_res[0]:.4f}, p-value: {adf_res[1]:.4f}")
    if adf_res[1] < 0.05:
        print(" Reject H0: Series is likely stationary.")
    else:
        print(" Fail to reject H0: Series is likely non-stationary.")

    try:
        kpss_res = kpss(series.dropna(), regression='c', nlags="auto")
        print(f"KPSS Statistic: {kpss_res[0]:.4f}, p-value: {kpss_res[1]:.4f}")
        if kpss_res[1] < 0.05:
            print(" Reject H0: Series is likely non-stationary.")
        else:
            print(" Fail to reject H0: Series is likely stationary.")
    except Exception as e:
        print("KPSS test error:", e)

stationarity_tests(ts1, "Foodstuff CPI")
stationarity_tests(ts3, "CPI Rate")
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(2, 2, figsize=(12,6))
plot_acf(ts1.dropna(), ax=ax[0,0]); ax[0,0].set_title("ACF - Foodstuff CPI")
plot_pacf(ts1.dropna(), ax=ax[0,1]); ax[0,1].set_title("PACF - Foodstuff CPI")
plot_acf(ts3.dropna(), ax=ax[1,0]); ax[1,0].set_title("ACF - CPI Rate")
plot_pacf(ts3.dropna(), ax=ax[1,1]); ax[1,1].set_title("PACF - CPI Rate")
plt.tight_layout(); plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import itertools

# --- Grid search over (p,d,q) with AIC/BIC comparison ---
p = d = q = range(0, 3)
pdq = list(itertools.product(p, d, q))

results = []
for param in pdq:
    try:
        model = ARIMA(ts1, order=param)
        res = model.fit()
        results.append((param, res.aic, res.bic))
    except:
        continue

res_df = pd.DataFrame(results, columns=['(p,d,q)', 'AIC', 'BIC']).sort_values('AIC').reset_index(drop=True)
print(res_df.head(10))

best_order = res_df.iloc[0]['(p,d,q)']
print(f"Best ARIMA order (by AIC): {best_order}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Split train/test (80/20 by time)
split = int(len(ts1)*0.8)
train, test = ts1[:split], ts1[split:]

model = ARIMA(train, order=best_order)
res = model.fit()

forecast = res.forecast(steps=len(test))
rmse = mean_squared_error(test, forecast, squared=False)
mae = mean_absolute_error(test, forecast)

print(f"RMSE: {rmse:.4f}, MAE: {mae:.4f}")

plt.figure(figsize=(10,4))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(forecast, label='Forecast', color='red')
plt.legend(); plt.title(f"ARIMA{best_order} Forecast vs Actual"); plt.show()
```

```python
import seaborn as sns
residuals = res.resid

fig, ax = plt.subplots(1,3, figsize=(14,3))
sns.histplot(residuals, ax=ax[0], kde=True)
ax[0].set_title("Residual Distribution")
plot_acf(residuals, ax=ax[1])
plot_pacf(residuals, ax=ax[2])
plt.tight_layout(); plt.show()

from statsmodels.stats.diagnostic import acorr_ljungbox
ljung = acorr_ljungbox(residuals, lags=[10], return_df=True)
print("Ljung-Box test p-value:", ljung['lb_pvalue'].iloc[0])
```

```python
import warnings
warnings.filterwarnings("ignore")
from statsmodels.tsa.statespace.sarimax import SARIMAX

P = D = Q = range(0, 2)
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in itertools.product(P, D, Q)]

best_aic = np.inf
best_param = None

for param in pdq:
    for sparam in seasonal_pdq:
        try:
            mod = SARIMAX(ts1, order=param, seasonal_order=sparam)
            res = mod.fit(disp=False)
            if res.aic < best_aic:
                best_aic = res.aic
                best_param = (param, sparam)
        except:
            continue

print("Best SARIMA parameters:", best_param, "AIC:", best_aic)

sarima_mod = SARIMAX(train, order=best_param[0], seasonal_order=best_param[1]).fit()
sarima_forecast = sarima_mod.forecast(steps=len(test))

sarima_rmse = mean_squared_error(test, sarima_forecast, squared=False)
print("SARIMA RMSE:", sarima_rmse)
```
